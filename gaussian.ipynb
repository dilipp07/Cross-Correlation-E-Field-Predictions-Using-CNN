```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
c = 3e+8  # Speed of light
lambda_ = 790e-9  # Wavelength
freq = c / lambda_  # Actual Frequency of light
freq1 =c / 680e-9 
T1= 1 / freq  # Period
T2 = 1 / freq1  # Period
width = 100e-15  # Full width at half maximum


# tstep =  T / 10000  # Time step

t1 = np.linspace(-100 * T1, 100 * T1, 10000000)  # Time range -100T to 100T
t2 = np.linspace(-100 * T2 , 100 * T2, 10000000)  # Time range -100T to 100T
```

```python
Egauss = np.exp(1j * (2 * np.pi * freq * t1)) * np.exp(-2 * np.log(2) * t1**2 / width**2)  # Gaussian intensity pulse
# The complex exponential in Python is represented as 1j for imaginary unit 'i'
envelop = np.exp(-2 * np.log(2) * t1**2 / width**2)  # Gaussian intensity envelope
```

```python
plt.figure()
plt.plot(t1 , np.real(Egauss), 'r', linewidth=0.05)
plt.plot(t1 , envelop, 'b', linewidth=2)
plt.xlabel('time (fs)')
plt.ylabel('Amplitude')
plt.legend(['Amplitude', 'Envelope'])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
